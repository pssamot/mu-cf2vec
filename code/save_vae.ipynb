```python
import os
import shutil
import sys

import numpy as np
import pandas as pd
import json
```

```python
# Just choose the name of the dataset directory
dataset  = 'ml-20m'
DATA_DIR = '/Users/tomas/Documents/FEUP/Tese/data/' + dataset
PARSE_DATA_DIR = os.path.join(DATA_DIR, 'processed_70_10_20/embeddings/vae/')
```

```python
embeddings = pd.read_csv(os.path.join(PARSE_DATA_DIR, 'parsed_embeddings.csv'),header=None)
embeddings = embeddings.drop([0], axis=1)
```

```python
embeddings.head()
```

```python
with open(os.path.join(PARSE_DATA_DIR, 'profile2id.json')) as f:
  profile2id = json.load(f)
```

```python
unique_uid = pd.DataFrame(data=profile2id.keys(), columns=['uid'])
```

```python
unique_uid.head()
```

```python
# Merge two Dataframes on index of both the dataframes
final_vae = unique_uid.merge(embeddings, left_index=True, right_index=True)
```

```python
final_vae.head()
```

```python
final_vae.to_csv(os.path.join(PARSE_DATA_DIR, 'vae_embeddings.csv'), index=False)
```
